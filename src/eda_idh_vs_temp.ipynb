```python
import pandas as pd


dfPovertyData = pd.read_csv("idh.csv", sep=",")
dfPovertyData
```

```python
df_subset = dfPovertyData.iloc[0:1, [0] + list(range(43, 53))]
df_subset
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Se df_subset jÃ¡ possui um nome para a coluna 0, substitua 'NomeColuna0' pelo nome real da coluna
data = df_subset.values[0][1:]  # ObtÃ©m os valores da linha, excluindo o primeiro valor (que Ã© o paÃ­s ou identificaÃ§Ã£o)

anos = list(range(2012, 2012 + len(data)))  # Supondo que 2012 seja o ano inicial e queremos plotar os anos subsequentes

# Cria o grÃ¡fico
plt.figure(figsize=(8, 5))
plt.plot(anos, data, marker='o', linestyle='-')
plt.title('IDH ao longo dos anos')
plt.xlabel('Ano')
plt.ylabel('IDH')
plt.grid(True)
plt.show()
```
